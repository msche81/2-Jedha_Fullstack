```python
bitcoin_id = first_ten_coins[0]['id']

playload = {'vs_currency':'usd', 'days':'365'}

bitcoin_ohlc = requests.get(f'https://api.coingecko.com/api/v3/coins/{bitcoin_id}/ohlc', params=playload).json()
bitcoin_ohlc
```

```python
# Solution 1:
from datetime import datetime

# creation of a list which will stock the elements of bitcoin_ohlc but with the human-readable date
bitcoin_ohlc_dates = []

# iterates on each element of bitcoin_ohlc
for l in bitcoin_ohlc: 
  # creates a sublist to store the modified sub-elements 
  l_sublist = []
  # iterates on each sub-element of the elements of bitcoin_ohlc
  for elem in l:
    # will apply the transformations only if elem is an integer
    if type(elem) == int:
      # converts the Unix Timestamp to a string and remove the 3 last characters 
      str_elem = str(elem)[:-3] 
      # converts the str_elem to an integer
      int_elem = int(str_elem)
      # converts the Unix timestamp to a human-readable date
      l_sublist.append(datetime.utcfromtimestamp(int_elem).strftime('%Y-%m-%d'))
    else :
      # if the elem is not an integer, adds it to the l_sublist as it is
      l_sublist.append(elem)
  
  # adds the l_sublist to the the bitcoin_ohlc_dates list
  bitcoin_ohlc_dates.append(l_sublist)

bitcoin_ohlc_dates
```

```python
# Solution 2:

from datetime import datetime

# creation of a list which will stock the elements of bitcoin_ohlc but with the human-readable date
bitcoin_ohlc_rdates = []
# iterates on each element of bitcoin_ohlc
for elt in bitcoin_ohlc :
    # for each element in bitcoin_ohlc, this code checks each sub-element. If the sub-element is not an integer, it leaves it as-is. If it is an integer (which may represent a Unix timestamp), it converts it to a string, removes the last three characters, converts it back to an integer, and then converts it to a human-readable date
    ohlc_date_convert = map(lambda x: x if type(x) != int else datetime.utcfromtimestamp(int(str(x)[:-3])).strftime('%Y-%m-%d'), elt)
    # transforms the ohlc for a given converted timestamp to a list
    ohlc_elt = list(ohlc_date_convert)
    # adds each created list of ohlc for a given timestamp in the bitcoin_ohlc_rdates list
    bitcoin_ohlc_rdates.append(ohlc_elt)

bitcoin_ohlc_rdates
```

```python
# Solution 3

from datetime import datetime

# creation of a list which will stock the elements of bitcoin_ohlc but with the human-readable date
new_bitcoin_ohlc = []

for i in range(len(bitcoin_ohlc)) :
    # creates a sublist to store the modified sub-elements 
    l_sublist = []
    # "drops" the 3 last characters of the Unix timestamp (which are zeros)
    bitcoin_ohlc_new = bitcoin_ohlc[i][0] / 1000
    # converts the Unix timestamp to a human-readable date
    l_sublist.append(datetime.fromtimestamp(bitcoin_ohlc_new).strftime(('%Y-%m-%d')))

    # appends the remaining elements of the sublist without modification
    for j in range(1, len(bitcoin_ohlc[i])):
        l_sublist.append(bitcoin_ohlc[i][j])
    
    # adds the modified sublist to the new_bitcoin_ohlc list
    new_bitcoin_ohlc.append(l_sublist)

new_bitcoin_ohlc
```

```python
crypto_data = {}
coins_list = [i['id'] for i in first_ten_coins]

for coin in coins_list:

  try:
    response = requests.get(f'https://api.coingecko.com/api/v3/coins/{coin}/ohlc?vs_currency=usd&days=365').json()

    # creation of a list which will stock the elements of response but with the human-readable date
    coin_ohlc_rdates = []
    # iterates on each element of response
    for elt in response :
        # for each element in bitcoin_ohlc, this code checks each sub-element. If the sub-element is not an integer, it leaves it as-is. If it is an integer (which may represent a Unix timestamp), it converts it to a string, removes the last three characters, converts it back to an integer, and then converts it to a human-readable date
        ohlc_date_convert = map(lambda x: x if type(x) != int else datetime.utcfromtimestamp(int(str(x)[:-3])).strftime('%Y-%m-%d'), elt)
        # transforms the ohlc for a given converted timestamp to a list
        ohlc_elt = list(ohlc_date_convert)
        # adds each created list of ohlc for a given timestamp in the coin_ohlc_rdates list
        coin_ohlc_rdates.append(ohlc_elt)
   

    crypto_data[coin] = {}
    for l in coin_ohlc_rdates:
      date = l[0]
      l.remove(l[0])
      crypto_data[coin].update({date: l})


  except:
    print(f'Issue encountered while calling API for {coin}')

crypto_data
```
