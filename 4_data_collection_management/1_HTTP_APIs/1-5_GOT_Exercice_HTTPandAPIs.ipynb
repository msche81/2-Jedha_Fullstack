```python
r = requests.get("https://anapioficeandfire.com/api/characters/2")
# We can use r.json().keys() in order to get an array of keys names
print("r.json().keys() == ", r.json().keys())
data = pd.DataFrame(columns = r.json().keys())
data.head()
```

```python
# Liste pour stocker temporairement chaque DataFrame
temp_list = []

for i in range(1, 151):
    r = requests.get(f"https://anapioficeandfire.com/api/characters/{i}")
    temp_list.append(pd.DataFrame([r.json()]))  # Ajoute chaque réponse JSON sous forme de DataFrame
    print(f'{i} personnage done')

# Concaténation de tous les DataFrames stockés dans la liste
data = pd.concat(temp_list, ignore_index=True)

# Affichage des premières lignes du DataFrame résultant
data.head()
```
