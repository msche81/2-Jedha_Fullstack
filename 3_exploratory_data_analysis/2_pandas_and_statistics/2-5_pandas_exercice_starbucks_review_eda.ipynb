```python
import pandas as pd
```

```python
starbucks_reviews = pd.read_csv('src/starbucks_reviews.csv')
display(starbucks_reviews.head())
```

```python
print(starbucks_reviews.shape)
```

```python
starbucks_reviews.dtypes
```

```python
starbucks_reviews['date'] = pd.to_datetime(starbucks_reviews['date'])
starbucks_reviews.dtypes
```

```python
starbucks_reviews['year'] = starbucks_reviews['date'].dt.year
starbucks_reviews['month'] = starbucks_reviews['date'].dt.month
starbucks_reviews['day'] = starbucks_reviews['date'].dt.day
starbucks_reviews['weekday'] = starbucks_reviews['date'].dt.dayofweek
```

```python
starbucks_reviews.head()
```

```python
starbucks_reviews["Review"] = starbucks_reviews["Review"].str.lower()
```

```python
starbucks_reviews.head()
```

```python
starbucks_reviews["Review_contains_Starbucks"] = starbucks_reviews["Review"].str.count("starbucks")
```

```python
starbucks_reviews.head()
```

```python
starbucks_reviews["wordcount"] = starbucks_reviews["Review"].str.split(" ").apply(lambda x: len(x))
starbucks_reviews.head()
```

```python
starbucks_reviews.describe(include = 'all')
```
