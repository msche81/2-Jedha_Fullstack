```python
import numpy as np
import math
```

```python
# Example with numpy: pow and sqrt
print("4 squared: ", math.pow(4, 2))
print("2 power 5: ", math.pow(2, 5)) #“2 à la puissance 5” signifie que le nombre 2 est multiplié par lui-même 5 fois.
print("Square root of 25: ", np.sqrt(25)) # sqrt is not in the math module !
print("Cube root of 64: ", np.cbrt(64)) # neither is cbrt ! La racine cubique de 64 signifie trouver un nombre qui, lorsqu’il est multiplié par lui-même trois fois, donne 64
```

```python
# Exponential and logarithm
# Example with numpy
print("Exponential of 1: ", np.exp(1))
print("Exponential of 0: ", np.exp(0))
print("Natural logarithm of 10: ", np.log(10)) # "log" refers to the natural logarithm
print("Base 10 logarithm of 10: ", np.log10(10)) # "log10" refers to the logarithm of base 10
```

```python
# trigonometrics functions
# Examples with numpy
print("Cosinus of 0: ", np.cos(0))
print("Cosinus of pi: ", np.cos(np.pi))
print("Cosinus of pi/2: ", np.cos(np.pi/2).round(2))
print()
print("Sinus of 0: ", np.sin(0))
print("Sinus of pi: ", np.sin(np.pi).round(2))
print("Sinus of pi/2: ", np.sin(np.pi/2))
print()
print("Tan of 0: ", np.tan(0))
print("Tan of pi: ", np.tan(np.pi).round(2))
print("Tan of pi/2: ", np.tan(np.pi/2)) # this will diverge !
print()
```

```python
# random numbers
print("Sample of 30 random values drawn from a discrete uniform distribution on [0, 10]: ")
print(np.random.randint(0, 10, 30))
print()
print("Sample of 20 random values drawn from a discrete uniform distribution on [-2, 2]: ")
print(np.random.randint(-2, 2, 20))
print()
```

```python
# Distribution Uniforme Discrète
# Générer 10 nombres entiers aléatoires entre 1 et 6 (inclus)
result = np.random.randint(1, 7, size=10)
print(result)
```

```python
# Distribution Binomiale
# Générer des échantillons binomiaux
n = 10  # Nombre d'essais
p = 0.5  # Probabilité de succès
size = 10  # Nombre d'échantillons à générer

result = np.random.binomial(n, p, size)
print(result)
```

```python
# Distribution de Poisson
# Générer des échantillons de Poisson
lambda_param = 3  # Taux moyen d'événements
size = 10  # Nombre d'échantillons à générer

result = np.random.poisson(lambda_param, size)
print(result)
```

```python
# Probability densities for continuous variables
# Distribution Normale (Gaussienne)
import matplotlib.pyplot as plt

# Générer des échantillons de la distribution normale
mean = 0  # Moyenne
std_dev = 1  # Écart type
samples = np.random.normal(mean, std_dev, 1000)

# Tracer l'histogramme des échantillons
plt.hist(samples, bins=30, density=True)
plt.title('Distribution Normale')
plt.xlabel('Valeur')
plt.ylabel('Densité de Probabilité')
plt.show()
```

```python
# Distribution Uniforme Continue
# Générer des échantillons de la distribution uniforme
low = 0  # Borne inférieure
high = 1  # Borne supérieure
samples = np.random.uniform(low, high, 1000)

# Tracer l'histogramme des échantillons
plt.hist(samples, bins=30, density=True)
plt.title('Distribution Uniforme Continue')
plt.xlabel('Valeur')
plt.ylabel('Densité de Probabilité')
plt.show()
```

```python
# Distribution Exponentielle
# Générer des échantillons de la distribution exponentielle
lambda_param = 1  # Taux
samples = np.random.exponential(1/lambda_param, 1000)

# Tracer l'histogramme des échantillons
plt.hist(samples, bins=30, density=True)
plt.title('Distribution Exponentielle')
plt.xlabel('Valeur')
plt.ylabel('Densité de Probabilité')
plt.show()
```
